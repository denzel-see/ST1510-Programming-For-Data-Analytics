```python
import matplotlib.pyplot as plt
import numpy as np
arr=[]
# Importing data
data = np.genfromtxt('CA1data/government-recurrent-expenditure-on-education-per-student.csv', dtype=['i4','U50','i8'], delimiter=",", names=True)

#Simple Text-Based Analysis
for x in range(0,len(data)):
    arr.append(list(data[x]))
print("***Government Recurrent Expenditure on Education per Student***")
print(f"There are {np.shape(arr)[0]} rows and {np.shape(arr)[1]} columns\n\nThe names of the columns are:\n-{data.dtype.names[0]} <class: {data[data.dtype.names[0]].dtype}> isnumeric: {np.issubdtype(data[data.dtype.names[0]].dtype, np.number)}\n-{data.dtype.names[1]} <class: {data[data.dtype.names[1]].dtype}> isnumeric: {np.issubdtype(data[data.dtype.names[1]].dtype, np.number)}\n-{data.dtype.names[2]} <class: {data[data.dtype.names[2]].dtype}> isnumeric: {np.issubdtype(data[data.dtype.names[2]].dtype, np.number)}")
print(f"\n{len(np.unique(data[data.dtype.names[0]]))} unique values in {data.dtype.names[0]} column\n{len(np.unique(data[data.dtype.names[1]]))} unique values in {data.dtype.names[1]} column\n{len(np.unique(data[data.dtype.names[2]]))} unique values in {data.dtype.names[2]} column")
#Segregating out the specific years
data_year = data[np.isin(data['year'], [2020,2019,2018,2017,2016])]
data_year09= data[np.isin(data['year'], [2009,2008,2007,2006,2005])]
data_yearline = data[np.isin(data['year'],[2020,2019,2018,2017,2016,2015,2014,2013,2012])]

#Segregating out the specific education institution
data_year_school = data_year[np.isin(data_year['type_of_educational_institution'], ['Junior Colleges/Centralised Institutes', 'Secondary Schools' , 'Full-time Nitec / Higher Nitec courses'])]
data_year09_school = data_year09[np.isin(data_year09['type_of_educational_institution'], ['Junior Colleges/Centralised Institutes', 'Secondary Schools' , 'Institute of Technical Education'])]
data_line = data_yearline[np.isin(data_yearline['type_of_educational_institution'], ['Junior Colleges/Centralised Institutes', 'Secondary Schools' , 'Institute of Technical Education'])]

#Putting the data of different education institution into specific varaiables 
data_jcci=data_year_school[data_year_school['type_of_educational_institution']=='Junior Colleges/Centralised Institutes']
data_secondary=data_year_school[data_year_school['type_of_educational_institution']=='Secondary Schools']
data_ite=data_year_school[data_year_school['type_of_educational_institution']=='Full-time Nitec / Higher Nitec courses']

data09_jcci=data_year09_school[data_year09_school['type_of_educational_institution']=='Junior Colleges/Centralised Institutes']
data09_secondary=data_year09_school[data_year09_school['type_of_educational_institution']=='Secondary Schools']
data09_ite=data_year09_school[data_year09_school['type_of_educational_institution']=='Institute of Technical Education']
                     
dataline_jcci=data_yearline[data_yearline['type_of_educational_institution']=='Junior Colleges/Centralised Institutes']
dataline_secondary=data_yearline[data_yearline['type_of_educational_institution']=='Secondary Schools']
dataline_ite=data_yearline[data_yearline['type_of_educational_institution']=='Full-time Nitec / Higher Nitec courses']

#Taking only the number value from the data
value_jcci=data_jcci['expenditure_per_student']
value_secondary=data_secondary['expenditure_per_student']
value_ite=data_ite['expenditure_per_student']

value09_jcci=data09_jcci['expenditure_per_student']
value09_secondary=data09_secondary['expenditure_per_student']
value09_ite=data09_ite['expenditure_per_student']

valueline_jcci=dataline_jcci['expenditure_per_student']
valueline_secondary=dataline_secondary['expenditure_per_student']
valueline_ite=dataline_ite['expenditure_per_student']


#Line Chart
plt.figure(figsize=(12,8),facecolor='lightblue')

plt.plot(valueline_secondary,'--', label="Secondary Schools")
plt.plot(valueline_jcci,'-.', label="Junior Colleges/\nCentralised Institutes")
plt.plot(valueline_ite, '-', label="Full-time Nitec / \nHigher Nitec courses")

plt.title("Government Recurrent Expenditure on Education per Student/Year\n",font={'size':18})
plt.xlabel("Year",font={'size':14})
plt.ylabel("Expenditure on Education Per Student (S$)",font={'size':14})
plt.xticks(np.arange(9),('2012','2013','2014','2015','2016','2017','2018','2019','2020'))
plt.yticks(np.arange(9000,19000,1000))
plt.legend(loc='lower right', shadow=True)

# Pie Chart

avg_secondary=np.sum(value_secondary)/5
avg_jcci=np.sum(value_jcci)/5
avg_ite=np.sum(value_ite)/5

avg09_secondary=np.sum(value09_secondary)/5
avg09_jcci=np.sum(value09_jcci)/5
avg09_ite=np.sum(value09_ite)/5


ratio=np.array([avg_secondary,avg_jcci,avg_ite])
ratio09=np.array([avg09_secondary,avg09_jcci,avg09_ite])

labels=('Secondary Schools', 'Junior Colleges/\nCentralised Institutes','Full-time Nitec / Higher Nitec courses')
colors=('r','#2CCDBD','#FFFF1A')



plt.figure(figsize=(14,8), facecolor='lightgreen')

plt.suptitle("\nAverage Expenditure on Education per Student over 5 years\n",fontsize=18)

plt.subplot(121)
plt.title("2005-2009",font={'size':14})
plt.pie(ratio09,labels=labels,colors=colors,autopct='%1.1f%%')

plt.subplot(122)
plt.title("2016-2020",font={'size':14})
plt.pie(ratio,labels=labels,colors=colors,autopct='%1.1f%%')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
arrPri=[]
# Importing data
dataGov = np.genfromtxt('CA1data/government-recurrent-expenditure-on-education-per-student.csv', dtype=['i4','U50','i8'], delimiter=",", names=True)
dataPri = np.genfromtxt('CA1data/percentage-of-p1-cohort-eligible-for-secondary-school.csv', dtype=['i4','f'], delimiter=",", names=True)

#Simple Text-Based Analysis
for x in range(0,len(dataPri)):
    arrPri.append(list(dataPri[x]))
print("***Government Recurrent Expenditure on Education per Student***")
print(f"There are {np.shape(arr)[0]} rows and {np.shape(arr)[1]} columns\n\nThe names of the columns are:\n-{data.dtype.names[0]} <class: {data[data.dtype.names[0]].dtype}> isnumeric: {np.issubdtype(data[data.dtype.names[0]].dtype, np.number)}\n-{data.dtype.names[1]} <class: {data[data.dtype.names[1]].dtype}> isnumeric: {np.issubdtype(data[data.dtype.names[1]].dtype, np.number)}\n-{data.dtype.names[2]} <class: {data[data.dtype.names[2]].dtype}> isnumeric: {np.issubdtype(data[data.dtype.names[2]].dtype, np.number)}")
print(f"\n{len(np.unique(data[data.dtype.names[0]]))} unique values in {data.dtype.names[0]} column\n{len(np.unique(data[data.dtype.names[1]]))} unique values in {data.dtype.names[1]} column\n{len(np.unique(data[data.dtype.names[2]]))} unique values in {data.dtype.names[2]} column")

print("\n***Percentage of P1 Cohort Eligible for Secondary School***")
print(f"There are {np.shape(arrPri)[0]} rows and {np.shape(arrPri)[1]} columns\n\nThe names of the columns are:\n-{dataPri.dtype.names[0]} <class: {dataPri[dataPri.dtype.names[0]].dtype}> isnumeric: {np.issubdtype(dataPri[dataPri.dtype.names[0]].dtype, np.number)}\n-{dataPri.dtype.names[1]} <class: {dataPri[dataPri.dtype.names[1]].dtype}> isnumeric: {np.issubdtype(dataPri[dataPri.dtype.names[1]].dtype, np.number)}")
print(f"\n{len(np.unique(dataPri[dataPri.dtype.names[0]]))} unique values in {dataPri.dtype.names[0]} column\n{len(np.unique(dataPri[dataPri.dtype.names[1]]))} unique values in {dataPri.dtype.names[1]} column")

dataGov_year = dataGov[np.isin(dataGov['year'], [2020,2019,2018,2017,2016,2015,2014,2013,2012,2011,2010,2009,2008,2007,2006,2005,2004,2003,2002,2001])]
dataGov_school = dataGov_year[dataGov_year['type_of_educational_institution']=='Primary Schools']
value_gov = dataGov_school['expenditure_per_student']

value_pri = dataPri['percentage']

plt.figure(figsize=(12,8),facecolor='#FDFD96')
plt.title("Correlation Between the Expenditure Spent per Student vs\n the Percentage of P1 Cohort Eligible for Secondary School\n", font={"size": 20})
plt.xlabel("Expenditure spent per Primary school student",font={"size": 13})
plt.ylabel("Percentage of P1 Cohort Eligible for Secondary School",font={"size": 13})
plt.xticks(np.arange(3000,13000,1000))
plt.yticks(np.arange(97.4,98.8,0.2))
plt.ylim(97.3,98.65)
plt.scatter(value_gov,value_pri, color='g')
m,b = np.polyfit(value_gov, value_pri, deg=1)
plt.plot(value_gov, m*value_gov + b, 'r-')

plt.figure(figsize=(12,8),facecolor='#FF6961')
plt.title("\nExpenditure per Primary School Student over the years\n",font={"size": 20})
plt.xlabel("Year",font={"size": 13})
plt.ylabel("Expenditure per Primary School Student",font={"size": 13})
plt.xticks(np.arange(2001,2021,1))
plt.bar(np.arange(2001,2021),value_gov)


plt.show()
```
